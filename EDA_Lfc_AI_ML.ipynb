### Passenger distribution by Gender

Histogram Plot





```python
import matplotlib.pyplot as plt


plt.hist(df['Sex'], bins=3)
plt.xlabel('Sex')
plt.ylabel('Frequency')
plt.title('Distribution by gender')
plt.show()
```

## Box plot for Age

```python
fig, ax = plt.subplots(figsize=(6, 8))
ax.boxplot(df["Age"].dropna(), patch_artist=True)
ax.set_xticklabels(["Box plot"])
plt.show()
```

```python
df['Pclass'] = df['Pclass'].astype('object')
```

```python
plt.hist(df['Pclass'], bins=5)
plt.xlabel('Pclass')
plt.ylabel('Frequency')
plt.title('Distribution by Pclass')
plt.show()
```

```python
corr_df = df[['Age', 'Fare', 'Siblings/Spouses Aboard', 'Parents/Children Aboard']].corr()
corr_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(corr_df, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```

    *   **Fare and Parents/Children Aboard:** There is a weak positive correlation (0.22), suggesting a similar slight tendency for passengers with more parents/children to pay a higher fare.
